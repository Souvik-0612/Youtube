```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vectorize the list of numbers using Numpy it will help us for broadcasting also!
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([1, 2.8, 5.1, 6.8, 9.5, 11])  # y = 2x + 1
n = len(x)
```

# Last equation

$$ n a_0 + \left( \sum_{i = 0}^{n-1}x_i\right) a_1 = \sum_{i = 0}^{n-1}y_i$$

$$  a_0 \left(\sum_{i = 0}^{n-1}x_i\right) +  \left(\sum_{i = 0}^{n-1}x_i^2\right) a_1 = \sum_{i = 0}^{n-1}x_iy_i$$

```python
# AX = B
A = np.array([[n, np.sum(x)],
              [np.sum(x), np.sum(x**2)]])
B = np.array([np.sum(y), np.sum(x*y)])
a0, a1 = np.linalg.solve(A, B)
a1
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x, y, 'o', label='Data')
plt.plot(x, a0 + a1*x, label='Fitted line')
plt.legend()
plt.show()
```

# Goodness of fit
$$C = \sqrt{\frac{S_y - E}{S_y}}$$
Where, $S_y = \sum_i(y_i - \bar y)^2$ and $E = \sum_i(y_i - a_0 - a_1 x_i)^2$

```python
Sy = np.sum((y - np.mean(y))**2)
E = np.sum((y - a0 - a1*x)**2)
C = np.sqrt((Sy - E)/Sy)
C*100
```
